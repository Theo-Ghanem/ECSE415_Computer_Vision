## Part 2-YOLOv8 Object Detection on Montréal Streets

### 1.Using my cellphone I captured the following street scene in Montréal.
(Shot on iPhone 12 mini)

```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("image_street_1.jpg")

# Convert the image from BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

### 2.Implement the trained YOLOv8 object detection model to identify what are the types of objects  included  in  the  image  (such  as  person,  bicycle,  vehicle,  tree)  and  count  the number of each object.

```python
# Load a pretrained YOLO model
model = YOLO('yolov8n.pt')

# Perform object detection on the image using the model
results = model.predict(image)
result = results[0]

# Convert the result to an image
detected_image = result.plot()
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(detected_image)
ax[1].set_title('Image with Detected Objects')
ax[1].axis('off')
plt.show()
```

```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt

# **Assuming you have correctly mounted your Google Drive and verified the image path**
image_path = 'image_street_1.jpg'  # Replace with your actual path

# Load the image
image = cv2.imread(image_path)

# Check if image was read successfully
if image is None:
  print("Error: Image not found. Please check the path.")
else:
  # Convert the image from BGR to RGB (optional, depending on YOLO model requirements)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

  # Load a pretrained YOLO model
  model = YOLO('yolov8n.pt')

  # Perform object detection on the image using the model
  results = model.predict(image)
  result = results[0]

  # Convert the result to an image
  detected_image = result.plot()

  # Create the plot
  fig, ax = plt.subplots(1, 2, figsize=(20, 10))

  # Display original image
  ax[0].imshow(image)
  ax[0].set_title('Original Image')
  ax[0].axis('off')

  # Display image with detected objects
  ax[1].imshow(detected_image)
  ax[1].set_title('Image with Detected Objects')
  ax[1].axis('off')

  plt.show()
```
